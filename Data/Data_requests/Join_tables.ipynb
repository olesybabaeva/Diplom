Импортируем данные (excel-файлы) о поданных заявках на ремонт оборудования из разных документов. Затем произведем объединение и преобразование данных.

```python
import pandas as pd
import numpy as np
from datetime import date
from datetime import datetime
import time
```

```python
df_1 = pd.read_excel('1.xlsx')
df_1.info()
```

```python
df_2 = pd.read_excel('2.xlsx')
df_3 = pd.read_excel('3.xlsx')
df_4 = pd.read_excel('4.xlsx')
df_5 = pd.read_excel('5.xlsx')
df_6 = pd.read_excel('6.xlsx')
df_7 = pd.read_excel('7.xlsx')
df_8 = pd.read_excel('8.xlsx')
df_9 = pd.read_excel('9.xlsx')
```

```python
df = pd.concat([df_1, df_2, df_3, df_4, df_5, df_6, df_7, df_8, df_9], ignore_index=True, axis=0)
df
```

```python
df.info()
```

В столбце "Дата закрытия заявки", где пустые значения, подставим дату "сегодня".

```python
df["Дата закрытия заявки"].fillna(time.strftime("%Y-%m-%d"), inplace=True)
df["Дата закрытия заявки"]
```

Рассчитаем продолжительность ремонта.

```python
df['длительность ремонта'] = df['Дата закрытия заявки'] - df['Дата открытия']
df
```

Изменим регистр в столбцах

```python
df['Оборудование'] = df['Оборудование'].str.lower()
df['Описание неисправ.'] = df['Описание неисправ.'].str.lower()
df['Причина неисправности'] = df['Причина неисправности'].str.lower()
df['Принятые меры'] = df['Принятые меры'].str.lower()
df
```

Удалим дубликаты, если они имеются (было удалено 90 повторяющихся строк)

```python
df = df.drop_duplicates()
df
```

```python
df.info()
```

Удалим пустые значения по столбцу "Цех".

```python
df = df.dropna(subset=['Цех'])
df.info()
```

Удалим пустые значения по столбцу "Дата открытия".

```python
df = df.dropna(subset=['Дата открытия'])
df.info()
```

Осталось 25074 строки с данными.

```python
df.to_csv('requests.csv')
```
