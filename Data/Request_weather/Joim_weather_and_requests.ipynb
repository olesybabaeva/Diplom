Объединим данные о погоде и о поломках оборудования. Это необходимо сделать по столбцам "Дата открытия" и "data".

```python
import pandas as pd
```

```python
df_requests = pd.read_csv('requests.csv')
df_requests.info()
```

```python
df_weather = pd.read_csv('weather.csv')
df_weather.info()
```

```python
df_requests['Дата открытия'] = df_requests['Дата открытия'].astype('datetime64[ns]')
df_requests['Дата закрытия заявки'] = df_requests['Дата закрытия заявки'].astype('datetime64[ns]')
df_requests.info()
```

```python
df_weather['data'] = df_weather['data'].astype('datetime64[ns]')
df_weather.info()
```

```python
df_requests['data'] = df_requests['Дата открытия']
df_requests.info()
```

```python
df_merged = df_requests.merge(df_weather, on='data', how='outer')
df_merged
```

```python
df_merged = df_merged.drop(['Unnamed: 0_x', 'Unnamed: 0_y'], axis=1)
```

```python
df_merged.info()
```

Удалим данные, где отсутствуют температура.

```python
df_merged = df_merged.dropna(subset=['temp'])
df_merged.info()
```

```python
df_merged = df_merged.dropna(subset=['Цех'])
df_merged.info()
```

После всех преобразований, у нас осталось 10535 строк с данными о выходе из строя оборудования с дневной температурой.

```python
df_merged = df_merged.drop(['длительность ремонта'], axis=1)
df_merged.info()
```

```python
df_merged['длительность ремонта'] = df_merged['Дата закрытия заявки'] - df_merged['Дата открытия']
df_merged
```

```python
df_merged[['Категория оборудования', 'Инв. номер']] = df_merged['Инв. номер'].str.split('-', n=1, expand=True)
df_merged
```

```python
df_merged.info()
```

```python
df_merged.to_csv('join.csv')
```
